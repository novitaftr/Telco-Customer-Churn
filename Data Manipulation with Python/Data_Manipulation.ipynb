**No 1**

```python
#tenure TIDAK samadengan 0
syarat = df['tenure'] != 0
df_baru = df[syarat]
df_baru.shape
```

**No 2**

```python
#TIDAK MEMILIKI PARTNER dan TIDAK MEMILIKI DEPENDENTS
syarat_1 = df_baru['Partner'] == 'No'
syarat_2 = df_baru['Dependents'] == 'No'

df_filter_1 = df_baru[syarat_1 & syarat_2]
df_filter_1
```

```python
df_filter_1['tenure'].mean()
```

**No 3**

```python
#MEMILIKI PARTNER dan TIDAK MEMILIKI DEPENDENTS
syarat_1 = df_baru['Partner'] == 'Yes'
syarat_2 = df_baru['Dependents'] == 'No'

df_filter_2 = df_baru[ syarat_1 & syarat_2 ]
df_filter_2
```

```python
df_filter_2['MonthlyCharges'].max()
```

**No 4**

```python
#TIDAK MEMILIKI PARTNER namun MEMILIKI DEPENDENTS
syarat_1 = df_baru['Partner'] == 'No'
syarat_2 = df_baru['Dependents'] == 'Yes'

df_filter_3 = df_baru[ syarat_1 & syarat_2 ]
df_filter_3
```

```python
df_filter_3['customerID'].count()
```

```python
#Senior Citizen = 0, memiliki PHONE SERVICE dan memiliki INTERNET SERVICE FIBER OPTIC
syarat_1 = df_baru['SeniorCitizen'] == 0
syarat_2 = df_baru['PhoneService'] == 'Yes'
syarat_3 = df_baru['InternetService'] == 'Fiber optic'

df_filter_4 = df_baru[syarat_1 & syarat_2 & syarat_3]
df_filter_4
```
